```python
# NOTE: This code requires the alpaca-py library to be installed.
# The exact import paths are inferred based on the documentation structure 
# (e.g., CryptoHistoricalDataClient is found under Historical Data [2, 3]).

from alpaca.data.historical.crypto import CryptoHistoricalDataClient
from alpaca.data.requests import CryptoLatestTradeRequest
# Although 'Trade' is referenced in the return type [1], 
# it may need to be imported depending on how you wish to process the data.

# --- Configuration (Authentication is optional but recommended) ---
# The CryptoHistoricalDataClient does not require authentication to use, 
# but authenticating increases your data rate limit [4].
# Replace these placeholders with your actual Alpaca API keys if desired:
# API_KEY = "YOUR_API_KEY"
# SECRET_KEY = "YOUR_SECRET_KEY"

# --- 1. Initialize the Crypto Historical Data Client ---
# Instantiating the client without keys, as authentication is optional [4].
client = CryptoHistoricalDataClient() 
    # If using keys, initialize as: client = CryptoHistoricalDataClient(api_key=API_KEY, secret_key=SECRET_KEY)

# --- 2. Define the Request ---
# We use CryptoLatestTradeRequest to specify the symbol we want data for [5, 6].
SYMBOL = "BTC/USD"
request_params = CryptoLatestTradeRequest(
    symbol_or_symbols=[SYMBOL] # The ticker identifier or list of ticker identifiers [5, 6].
)

# --- 3. Retrieve the Latest Trade Data ---
# get_crypto_latest_trade returns the latest trade for the coin(s) [1].
# The function returns a dictionary mapping the symbol string to the Trade object [1, 7].
latest_trade_result = client.get_crypto_latest_trade(request_params=request_params)

print(f"{SYMBOL}: {latest_trade_result}")
 
```

```python
# ... existing code ...
from time import sleep
from alpaca.data.historical.crypto import CryptoHistoricalDataClient
from alpaca.data.requests import CryptoSnapshotRequest

SYMBOL = "ETH/USD"
client = CryptoHistoricalDataClient()
request_params = CryptoSnapshotRequest(
    symbol_or_symbols=[SYMBOL]
)
try:
    for i in range(10):
        print(f"Fetching snapshot data for {SYMBOL}...")
        snapshot_data = client.get_crypto_snapshot(request_params=request_params)
        if snapshot_data and SYMBOL in snapshot_data:
            snapshot = snapshot_data[SYMBOL]
            print("\n--- Snapshot Details ---")
            print(f"Symbol: {SYMBOL}")
            if hasattr(snapshot, 'latest_trade') and snapshot.latest_trade:
                print(f"Latest Trade Price: {snapshot.latest_trade.price}")
            if hasattr(snapshot, 'latest_quote') and snapshot.latest_quote:
                print(f"Latest Quote (Bid/Ask): {snapshot.latest_quote.bid_price} / {snapshot.latest_quote.ask_price}")
            if hasattr(snapshot, 'latest_minute_bar') and snapshot.latest_minute_bar:
                print(f"Latest Minute Bar Close: {snapshot.latest_minute_bar.close}")
            if hasattr(snapshot, 'latest_daily_bar') and snapshot.latest_daily_bar:
                print(f"Latest Daily Bar Close: {snapshot.latest_daily_bar.close}")
            if hasattr(snapshot, 'previous_daily_bar') and snapshot.previous_daily_bar:
                print(f"Previous Daily Bar Close: {snapshot.previous_daily_bar.close}")
            sleep(60)
        else:
            print(f"Could not retrieve snapshot data for {SYMBOL}.")
except Exception as e:
    print(f"An error occurred while fetching crypto snapshot: {e}")
```
